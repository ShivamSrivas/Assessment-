````python
def normalize_strings(data):
    return [item.strip().lower() for item in data if isinstance(item, str)]
"""
It's a function where the expected parameter is an iterable object. It loops through the argument, checks if each element 
is a string, and if it is, remove string with leading and trailing whitespace and converts it to lowercase and appends it to a list using list comprehension.

                                                 *******flow***********

for loop --1--> <--2-- if condition false
for loop --1--> if condition true --2--> 'remove string with leading and trailing whitespace and converts it to lowercase and appends'
"""
````

````python
def calculate_averages(values):
    numbers = [num for num in values if isinstance(num, (int, float))]
    if not numbers:
        raise ValueError("No valid numbers found")
    return sum(numbers) / len(numbers)
"""
It's a function where the expected parameter is an iterable object. It loops through the argument, checks if each element 
is an integer or float, and if it is, appends it to a list using list comprehension, which is saved in a variable called 'numbers'. 
If 'numbers' is empty, it raises a 'No valid numbers found' error else return add all numbers in list and divide by total number of elements.

                                                 *******flow***********

for loop --1--> <--2-- if condition false
for loop --1--> if condition true --2--> 'if it's a number append it' --3--> if false --4--> raise ValueError("No valid numbers found")
for loop --1--> if condition true --2--> 'if it's a number append it' --3--> if true --4--> return add all numbers in list and divide by total number of elements
"""
````

````python
def apply_tax(prices, tax_rate):
    return [price * (1 + tax_rate) for price in prices]

"""
It's a function that expects two parameters: one as an iterable object and the second as a number (int or float). 
In the list comprehension, it loops through the iterable object named 'prices', and for every element, it adds 1 to the 
tax_rate, multiplies it with the price, appends the result to a list, and returns the list.

                                                *******flow***********

for loop --1--> adds 1 to the tax_rate --2--> multiplies it with the price --3--> appends the result to a list --4--> returns list
"""
````

````python
def filter_above_threshold(data, threshold):
    return {k: v for k, v in data.items() if v > threshold}

"""
It's a function that expects two parameters: one as a dictionary and the second as a number (int or float). 
It then loops through the key-value pairs in the dictionary and checks if the value is greater than the threshold. 
If yes, it adds the key-value pair to the resulting dictionary using dictionary comprehension.

                                              *******flow***********
                                              
for loop --1--> <--2-- if false
for loop --1--> if true --2--> append in dict --3--> return dict
"""
````

````python
def count_occurrences(items):
    counts = {}
    for item in items:
        counts[item] = counts.get(item, 0) + 1
    return counts

"""
It's a function that expects a parameters which will be iterable object and then it will be counting occurrences of each element,
and appending in dictonary with key as element and value as total number of ocurrence

                                            *******flow***********
                                            
for loop --1--> count_ocurrence --2--> append in dict count 
"""
````

````python
def update_inventory(inventory, updates):
    for item, quantity in updates.items():
        if item in inventory:
            inventory[item] += quantity
        else:
            inventory[item] = quantity
    return inventory

"""
It's a function that expects two parameters: both as a dictionary name followed as inventory and updates . 
It then loops through the key-value pairs in the updates.keys and see if the key is in inventory or not if it is
then add quantity which is value of updates with inventory values else add new key values in inventory. 

                                            *******flow***********
                                            
for loop --1--> if false --2--> inventory[item] = quantity
for loop --1--> if false --2--> inventory[item] += quantity 
"""
````
